```python
import pandas as pd
from pandas_datareader import data as web
from datetime import date
```

```python
# Puxando a cotação

cotacao = web.DataReader('CSNA3.SA', data_source='yahoo', start='01-01-2000', end='08-31-2022')['Adj Close']
cotacao = cotacao.to_frame()
cotacao.reset_index(inplace=True)
cotacao.rename(columns={'Date':'Data'}, inplace=True)
cotacao
```

### Agora eu quero saber qual foi a última cotação de cada mês

Vou tentar adcionar uma coluna que retorna o nome do mês

```python
cotacao = cotacao.assign(Meses = 0)
cotacao['Meses'] = cotacao['Data'].dt.month
cotacao['Ano'] = cotacao['Data'].dt.year

cotacao
```

### Agora eu tenho que calcular  o último dia útil de cada mês


* Inicialmente vou tentar com um loop, para ir fixando e se conseguir melhor faço

```python
meses = cotacao['Meses']
```

```python
c = 1
lista_dias = []

for loop in range(0, len(meses)-1):
    atual = meses[c]
    anterior = meses[c-1]
    
    print(f'Iteraçao {c}')
    print(f'Atual: {atual}')
    print(f'Anterior: {anterior}')
    print(f'-----------------------')

    if atual != anterior:
        lista_dias.append(c-1)
        print('\33[96m Troca****************************************')
        print(cotacao['Data'][c-1])
    c += 1
```

```python
# Utilizando os valores do index, para selecionar apenas os dias desejados

cotacao_final = pd.DataFrame()

lista_datas = list()
lista_cotacoes = list()
lista_referencias = list()



for i in lista_dias: 
    lista_datas.append(cotacao.iloc[i]['Data'])
    lista_cotacoes.append(cotacao.iloc[i]['Adj Close'])

    ano = cotacao.iloc[i]['Ano']
    mes = cotacao.iloc[i]['Meses']
    lista_referencias.append(str(f'{mes}-{ano}'))
```

```python
cotacao_final['Data'] = lista_datas
cotacao_final['Cotacoes'] = lista_cotacoes
cotacao_final['Ref'] = lista_referencias
cotacao_final
```

```python
cotacao_final.to_csv('cotacao.csv')
```
